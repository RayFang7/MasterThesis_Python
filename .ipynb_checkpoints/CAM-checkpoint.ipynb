```python
import os
import PIL
import numpy as np
import torch
import torch.nn.functional as F
import torchvision.models as models
from torchvision import transforms
from torchvision.utils import make_grid, save_image
from gradcam.utils import visualize_cam
from gradcam import GradCAM, GradCAMpp
import nets
from lib import get_datatransformer
from arg import get_params
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
model = nets.vggnet("")
model.load_state_dict(torch.load("state_dict/210129-1712_2.dict"))
img_path = '/mnt/ramdisk/317/ap/1/1_H/Healthy control_patient105_no_ap_s0303lre5.jpg'
#img_path = "/home/ray/TrainingData/410/ap/10/1.MI/514_ILMI_p12005_ap10.jpg"
#img_path = '/mnt/ramdisk/419/ap/10/1_H/59_NORM_p19475_ap1.jpg'
args = vars(get_params())
transformer = get_datatransformer(args,2)
```

```python
model = nets.vggnet("")
model.load_state_dict(torch.load("temp.dict"))
img_path = '/home/ray/TrainingData/320/ap/5/0_MI/MI_patient004_antero-septal_ap_s0020are4.jpg'
#img_path = '/mnt/ramdisk/419/ap/10/1_H/59_NORM_p19475_ap1.jpg'
args = vars(get_params())
transformer = get_datatransformer(args,2)
```

```python
pil_img = PIL.Image.open(img_path)
torch_img = transformer['val'](pil_img).to(device)
normed_torch_img = torch_img.unsqueeze(0)
```

```python
configs = [
    dict(model_type='vgg', arch=model, layer_name='features_52'),
]
```

```python
for config in configs:
    config['arch'].to(device).eval()

cams = [
    [cls.from_config(**config) for cls in (GradCAM, GradCAMpp)]
    for config in configs
]
```

```python
images = []
for gradcam, gradcam_pp in cams:
    mask, _ = gradcam(normed_torch_img)
    heatmap, result = visualize_cam(mask, torch_img)

    mask_pp, _ = gradcam_pp(normed_torch_img)
    heatmap_pp, result_pp = visualize_cam(mask_pp, torch_img)
    
    images.extend([torch_img.cpu(), heatmap, heatmap_pp, result, result_pp])
    
grid_image = make_grid(images, nrow=5)
```

```python
transforms.ToPILImage()(grid_image)
```
